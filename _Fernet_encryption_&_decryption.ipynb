```python
import pandas as pd
from cryptography.fernet import Fernet
import os
from google.colab import files # Import files for file upload

# Generate a key and store it (only do this once)
key = Fernet.generate_key()
with open('secret.key', 'wb') as key_file:
    key_file.write(key)

# Load the key
def load_key():
    return open('secret.key', 'rb').read()

# Encrypt the data
def encrypt_data(data, key):
    fernet = Fernet(key)
    encrypted_data = fernet.encrypt(data.encode())
    return encrypted_data

# Upload the Excel file instead of using a local path
uploaded = files.upload() # This will prompt you to upload the file
file_path = next(iter(uploaded)) # Get the name of the uploaded file

# Load the Excel file
#file_path = 'C:/Users/theha/Downloads/cancer patient data sets.xlsx'  # Google Colab specific path
df = pd.read_excel(file_path)

# Load the encryption key
key = load_key()

# Create a directory to store the encrypted files
output_dir = 'encrypted_files/'
os.makedirs(output_dir, exist_ok=True)

# Encrypt each patient's data and save it in the specified directory
for index, row in df.iterrows():
    patient_id = row['Patient Id']
    patient_data = row.to_string()

    # Encrypt the patient's data
    encrypted_data = encrypt_data(patient_data, key)

    # Save encrypted data to the specified directory
    file_name = os.path.join(output_dir, f"patient_{patient_id}.enc")
    with open(file_name, 'wb') as enc_file:
        enc_file.write(encrypted_data)

# Zip the encrypted files for easy download
!zip -r encrypted_files.zip encrypted_files/

# Download the zip file
from google.colab import files
files.download('encrypted_files.zip')
```

```python
import zipfile
import os
from cryptography.fernet import Fernet

# Upload the ZIP file containing encrypted files
from google.colab import files
uploaded = files.upload()  # Upload the .zip file

# Extract the uploaded ZIP file
zip_file_name = next(iter(uploaded))  # Get the uploaded file name
with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extractall('/content/encrypted_files/')  # Extract to this directory

# Load the encryption key (you should have this from the previous step)
def load_key():
    return open('secret.key', 'rb').read()

# Decrypt the data
def decrypt_data(encrypted_data, key):
    fernet = Fernet(key)
    decrypted_data = fernet.decrypt(encrypted_data)
    return decrypted_data.decode()

# Load the encryption key
key = load_key()

# Directory containing the encrypted files
encrypted_files_dir = '/content/encrypted_files/'

# Decrypt each file and print the decrypted patient data
for file_name in os.listdir(encrypted_files_dir):
    if file_name.endswith('.enc'):  # Only process encrypted files
        file_path = os.path.join(encrypted_files_dir, file_name)

        # Read the encrypted data from the file
        with open(file_path, 'rb') as enc_file:
            encrypted_data = enc_file.read()

        # Decrypt the data
        decrypted_data = decrypt_data(encrypted_data, key)

        # Display the decrypted patient data
        print(f"Decrypted data from {file_name}:")
        print(decrypted_data)
        print("\n" + "="*50 + "\n")

# Optionally, you can save the decrypted files
decrypted_dir = '/content/decrypted_files/'
os.makedirs(decrypted_dir, exist_ok=True)

# Save decrypted data as text files
for file_name in os.listdir(encrypted_files_dir):
    if file_name.endswith('.enc'):
        file_path = os.path.join(encrypted_files_dir, file_name)

        with open(file_path, 'rb') as enc_file:
            encrypted_data = enc_file.read()

        decrypted_data = decrypt_data(encrypted_data, key)

        # Save to a new text file
        output_file_name = file_name.replace('.enc', '.txt')
        with open(os.path.join(decrypted_dir, output_file_name), 'w') as dec_file:
            dec_file.write(decrypted_data)

# Zip the decrypted files for downloading
!zip -r decrypted_files.zip /content/decrypted_files/

# Download the decrypted files
files.download('decrypted_files.zip')
```
